```python
my_stats.isna().sum()
new_df=my_stats[(my_stats['home_team']=='Arsenal') & (my_stats['result']=='H')]



```

```python
#Which team among the bix six has won the maximum matches from 2006/07 to 2017/2018
Bix_Six=['Arsenal','Tottenham Hotspur','Manchester United','Manchester City','Chelsea','Liverpool']
total={}
for team in Bix_Six:
    new_df_home=my_stats[(my_stats['home_team']==team) & (my_stats['result']=='H')]
    new_df_away=my_stats[(my_stats['away_team']==team) & (my_stats['result']=='A')]
    total_wins=len(new_df_home) + len(new_df_away)
    total[team]=total_wins
    
    
```

```python
#Plot winnings!
Teams=total.keys()
Winnings=total.values()
explode=[0,0,0.5,0,0,0]
plt.pie(Winnings,labels=Teams,autopct='%.2f%%',explode=explode,shadow=True)
plt.title('Winnings among Bix Six')
plt.show()
```

```python
# Which team among the bix six has lost maximum matches from 2006/07 to 2017/2018
Bix_Six=['Arsenal','Tottenham Hotspur','Manchester United','Manchester City','Chelsea','Liverpool']
total_lost={}
for team in Bix_Six:
    new_df_home=my_stats[(my_stats['home_team']==team) & (my_stats['result']=='A')]
    new_df_away=my_stats[(my_stats['away_team']==team) & (my_stats['result']=='H')]
    total_wins=len(new_df_home) + len(new_df_away)
    total_lost[team]=total_wins
```

```python
Teams=total_lost.keys()
Winnings=total_lost.values()
plt.bar(Teams,Winnings,color='white',edgecolor='Red')
plt.xticks(rotation=90)
plt.grid()
ax=plt.axes()
ax.set_facecolor('grey')
plt.xlabel('The Big Six',)
plt.ylabel('No of matches Lost')
plt.title('Most No of Matches Lost')
plt.show()
```

```python
#Most goals scored by teams in Big Six from 2006/07 to 2017/18
total_goals={}
for team in Bix_Six:
    goal=0
    total_goals[team]=goal
    h=my_stats[my_stats['home_team']==team]
    g=h['home_goals'].sum()
    a=my_stats[my_stats['away_team']==team]
    g_a=a['away_goals'].sum()
    tot_goal=g+g_a
    total_goals[team]=tot_goal
    
```

```python
teams=list(total_goals.keys())
winnings=list(total_goals.values())
x_teams=[i for i,_ in enumerate(teams)]
plt.barh(x_teams,winnings,color='grey')
plt.xlabel('No of Goals Scored')
ax=plt.axes()
ax.set_facecolor('white')
plt.ylabel('The Big Six')
plt.title('Most No of Goals')
plt.yticks(x_teams,teams)
plt.grid()
plt.show()
```

```python
season_groups=my_stats.groupby(['season'])
```

```python
season_wise_goals_HOME={}
season_wise_goals_AWAY={}
home_total_goals={}
away_total_goals={}
for name,group in season_groups:
    for team in Bix_Six:
        away_goals=group[group['away_team']==team]
        a_goals=away_goals['away_goals'].sum()
        home_goals=group[group['home_team']==team]
        h_goals=home_goals['home_goals'].sum()
        home_total_goals[team]=h_goals
        away_total_goals[team]=a_goals
    season_wise_goals_HOME[name]=home_total_goals
    season_wise_goals_AWAY[name]=away_total_goals
    home_total_goals={}
    away_total_goals={}
```

```python
#Home and Away Goals by Big Six for Season 2006/2007
season1_h=season_wise_goals_HOME['2006-2007']
season1_a=season_wise_goals_AWAY['2006-2007']

teams=list(season1_h.keys())
goals_h=list(season1_h.values())
goals_a=list(season1_a.values())
width=0.35
n=np.arange(6)
```

```python
plt.bar(n,goals_h,width,label='home')
plt.bar(n+width,goals_a,width,label='away')
plt.xticks(n + width / 3, ('Arsenal', 'Tottenham Hotspur', 'Manchester United', 'Manchester City', 'Chelsea','Liverpool'))
plt.legend()
plt.xticks(rotation=90)
plt.ylabel('Goals')
plt.title("Home and Away goals for 'The Big Six' for Season 2006/2007")
plt.grid()
plt.show()
```
